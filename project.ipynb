```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
order = pd.read_csv("data.csv")
order.head(50)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='excel_marks', y='python_marks', data=order)
```

```python
df = pd.DataFrame(order)

df['total_marks'] = df[['sql_marks','excel_marks','python_marks','power_bi_marks','english_marks']].sum(axis=1)


location_scores = df.groupby('location')['total_marks'].mean().reset_index().sort_values(by='total_marks', ascending=False)

print("Average Total Marks by Location:")
print(location_scores)

plt.figure(figsize=(7,7))
plt.pie(location_scores['total_marks'], labels=location_scores['location'], autopct='%1.1f%%', startangle=140)
plt.title('Average Total Marks by Location')
plt.show()
```
